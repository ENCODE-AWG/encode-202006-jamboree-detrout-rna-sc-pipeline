```python
import os
from woldrnaseq.models import load_gtf_cache
```

```python
gtf = load_gtf_cache(os.path.expanduser('~/proj/genome/mm10-M21-male/mm10-M21-male.h5'))
```

```python
genes = gtf[(gtf.type.isin(['transcript', 'tRNA'])) | (gtf.source == 'spikein')][['transcript_id', 'gene_id']]
genes.shape
```

```python
with open('genome/mm10-M21-male/txp2gene.tsv', 'wt') as outstream:
    for i, row in genes.iterrows():
        outstream.write(f"{row.transcript_id}\t{row.gene_id}\n")
```
