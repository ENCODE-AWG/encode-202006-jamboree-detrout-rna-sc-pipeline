```python
import os
from pathlib import Path
import time
import pysam
from xopen import xopen
```

```python
tenx_dir = Path("ENCSR874BOF_e10_5_limb/")
star_solo = tenx_dir / "diane_star_solo"
star_genome = star_solo / "Aligned.sortedByCoord.out.bam"
common_barcodes_filename = tenx_dir / "common_barcodes.txt"
```

```python
common_barcodes = set()
with open(common_barcodes_filename, 'r') as instream:
    for line in instream:
        common_barcodes.add(line.rstrip())
```

```python
len(common_barcodes)
```

```python
t_start  = time.time()
time.asctime()
```

```python
with xopen('ENCSR874BOF_common_reads.fastq.gz', 'wt') as outstream:
    with pysam.AlignmentFile(star_genome, 'rb') as alignments:
        count = 0
        for read in alignments:
            if read.has_tag('CB') and read.get_tag('CB') in common_barcodes:
                count += 1
                #print(count, read)
                outstream.write(f">{read.query_name}_{read.get_tag('CB')}\n{read.seq}\n+\n{read.qual}\n")

            if count % 100000 == 0:
                print(count)
```

```python
t_end = time.time()
time.asctime()
```
